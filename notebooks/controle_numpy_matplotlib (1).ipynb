# Partie C — Transformations Avancées avec NumPy

### C11. Espèces uniques
Listez les espèces uniques avec `np.unique` et comptez leurs occurrences.


```python
data = np.genfromtxt('Iris.csv', delimiter=',', skip_header=1, dtype=str, usecols=5)

# Calcul des espèces uniques et de leurs occurrences
especes, comptes = np.unique(data, return_counts=True)

# Affichage des résultats
for e, c in zip(especes, comptes):
    print(f"{e}: {c}")
```

### C12. Reshape par espèce
Transformez `features` en un array de forme `(3, 50, 4)` nommé `features_by_species`.


```python
# On saute l'en-tête et la colonne ID (index 0)
features = np.genfromtxt('Iris.csv', delimiter=',', skip_header=1, usecols=(1, 2, 3, 4))

# 2. Redimensionnement en (3 espèces, 50 échantillons, 4 caractéristiques)
features_by_species = features.reshape(3, 50, 4)

# Vérification de la forme
print(f"Nouvelle forme : {features_by_species.shape}")
```

### C13. Concaténation
Concaténez horizontalement les moyennes par colonne avec un array de zéros.


```python
# 1. Calcul des moyennes par colonne (axis=0) sur les 4 mesures
means = np.mean(features, axis=0)

# 2. Création d'un array de zéros de la même taille (ici 4 éléments)
zeros = np.zeros_like(means)

# 3. Concaténation horizontale
# Note : on utilise [means] et [zeros] pour s'assurer qu'ils sont traités comme une ligne
result = np.hstack((means, zeros))

print("Résultat de la concaténation :")
print(result)
```

### C14. Data cleaning (simulation)
Simulez des valeurs NaN et remplacez-les avec `np.nan_to_num`.


```python
# 1. Copie des données pour ne pas modifier l'original
features_with_nan = features.copy()

# 2. Simulation de NaN aux indices [0,0] et [1,2]
features_with_nan[0, 0] = np.nan
features_with_nan[1, 2] = np.nan

print("Données avec NaN (extraits) :\n", features_with_nan[:2])

# 3. Remplacement des NaN par 0 (ou une autre valeur choisie)
# nan_to_num remplace NaN par 0 et les infinis par de très grands nombres
features_cleaned = np.nan_to_num(features_with_nan, nan=0.0)

print("\nDonnées après np.nan_to_num (extraits) :\n", features_cleaned[:2])
```

### E22. Subplots comparatifs
Deux scatter plots côte à côte : Setosa vs Versicolor.


```python
# 1. Préparation des données (en utilisant features_by_species)
# Setosa est à l'index 0, Versicolor à l'index 1
setosa_petal_len = features_by_species[0, :, 2]
setosa_petal_wid = features_by_species[0, :, 3]

versicolor_petal_len = features_by_species[1, :, 2]
versicolor_petal_wid = features_by_species[1, :, 3]

# 2. Création des sous-graphiques (1 ligne, 2 colonnes)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

# Graphique 1 : Setosa
ax1.scatter(setosa_petal_len, setosa_petal_wid, color='purple', alpha=0.7)
ax1.set_title('Espèce : Setosa')
ax1.set_xlabel('Longueur pétale (cm)')
ax1.set_ylabel('Largeur pétale (cm)')

# Graphique 2 : Versicolor
ax2.scatter(versicolor_petal_len, versicolor_petal_wid, color='orange', alpha=0.7)
ax2.set_title('Espèce : Versicolor')
ax2.set_xlabel('Longueur pétale (cm)')

plt.tight_layout()
plt.show()
```

### E23. Grille 2x2
Histogrammes des 4 caractéristiques avec `plt.subplots(2,2)`.


```python
# Noms des colonnes pour les titres
titles = ['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width']
colors = ['#4C72B0', '#55A868', '#C44E52', '#8172B2']

# Création de la grille 2x2
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# aplatir la matrice axes (2,2) en un vecteur de 4 pour faciliter la boucle
for i, ax in enumerate(axes.flat):
    ax.hist(features[:, i], bins=15, color=colors[i], edgecolor='black', alpha=0.7)
    ax.set_title(titles[i])
    ax.set_xlabel('cm')
    ax.set_ylabel('Fréquence')
    ax.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

### E24. Bar chart avec error bars
Moyennes par espèce avec écart-type.


```python
# 1. Calcul des moyennes et écarts-types pour la 1ère colonne (SepalLength)
# axis=1 car on veut la moyenne des 50 échantillons pour chaque espèce
means = np.mean(features_by_species[:, :, 0], axis=1)
std_devs = np.std(features_by_species[:, :, 0], axis=1)

species_names = ['Setosa', 'Versicolor', 'Virginica']

# 2. Création du Bar Chart avec barres d'erreur (yerr)
plt.figure(figsize=(8, 6))
plt.bar(species_names, means, yerr=std_devs, capsize=7, color=['#4C72B0', '#55A868', '#C44E52'], alpha=0.8)

# 3. Personnalisation
plt.title("Moyenne de la longueur des sépales par espèce (+/- Écart-type)")
plt.ylabel("Longueur (cm)")
plt.grid(axis='y', linestyle='--', alpha=0.3)

plt.show()
```

### E25. Scatter coloré par espèce
Scatter plot avec couleurs par espèce.


```python
# Définition des espèces et des couleurs associées
species = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
colors = ['purple', 'orange', 'green']

plt.figure(figsize=(8, 6))

# Boucle pour tracer chaque espèce avec sa couleur et son label
for i in range(3):
    plt.scatter(features_by_species[i, :, 0],  # Longueur sépale
                features_by_species[i, :, 1],  # Largeur sépale
                color=colors[i],
                label=species[i],
                alpha=0.8,
                edgecolors='w')

# Personnalisation du graphique
plt.title("Dimensions des sépales par espèce")
plt.xlabel("Longueur du sépale (cm)")
plt.ylabel("Largeur du sépale (cm)")
plt.legend()  # Affiche la légende avec les noms d'espèces
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

### E26. Line plot superposé
Cumsum des largeurs des pétales par espèce.


```python
# 1. Extraction des largeurs de pétales (index 3 dans la dernière dimension)
# features_by_species a la forme (3, 50, 4)
setosa_cumsum = np.cumsum(features_by_species[0, :, 3])
versicolor_cumsum = np.cumsum(features_by_species[1, :, 3])
virginica_cumsum = np.cumsum(features_by_species[2, :, 3])

# 2. Création du graphique
plt.figure(figsize=(10, 6))
plt.plot(setosa_cumsum, label='Setosa', color='purple', linewidth=2)
plt.plot(versicolor_cumsum, label='Versicolor', color='orange', linewidth=2)
plt.plot(virginica_cumsum, label='Virginica', color='green', linewidth=2)

# 3. Personnalisation
plt.title("Somme cumulée des largeurs de pétales par espèce")
plt.xlabel("Nombre d'échantillons cumulés")
plt.ylabel("Somme cumulée de la largeur (cm)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

### F27. Corrélations
Calculez la matrice de corrélation avec `np.corrcoef`.


```python
correlation_matrix = np.corrcoef(features, rowvar=False)

# Affichage de la matrice
print("Matrice de corrélation (4x4) :")
print(correlation_matrix.round(2))
```

### F28. Visualisation des corrélations
Visualisez les corrélations sous forme de bar chart ou scatter.


```python
# On extrait les corrélations de la 4ème colonne (PetalWidth) avec les autres
# correlation_matrix a été calculée précédemment (F27)
target_corr = correlation_matrix[3, :3]
labels = ['Sepal Length', 'Sepal Width', 'Petal Length']

plt.figure(figsize=(8, 5))
plt.bar(labels, target_corr, color=['#4C72B0', '#55A868', '#C44E52'])

plt.axhline(0, color='black', linewidth=0.8) # Ligne à zéro
plt.title("Corrélation des caractéristiques avec la Largeur des Pétales")
plt.ylabel("Coefficient de Corrélation")
plt.ylim(-1, 1) # L'échelle d'une corrélation est toujours entre -1 et 1
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```

### F29. Variance par espèce
Identifiez l’espèce avec la plus grande variance globale et visualisez ses distributions.


```python
# Calcul de la variance pour chaque espèce et chaque colonne (axis=1)
# Ensuite, on somme les variances des 4 colonnes (axis=1 du résultat)
variances_par_espece = np.var(features_by_species, axis=1).sum(axis=1)

species_names = ['Setosa', 'Versicolor', 'Virginica']
idx_max = np.argmax(variances_par_espece)

print(f"Variances totales :")
for name, v in zip(species_names, variances_par_espece):
    print(f"- {name}: {v:.4f}")

print(f"\nL'espèce avec la plus grande variance globale est : {species_names[idx_max]}")
```
